```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
sigma_A = 0.15  # 15%
sigma_B = 0.20  # 20%
weights = np.linspace(0, 1, 100)
correlations = np.linspace(-1, 1, 100)

W, Rho = np.meshgrid(weights, correlations)
portfolio_variance = (W**2) * (sigma_A**2) + ((1 - W)**2) * (sigma_B**2) + \
                     2 * W * (1 - W) * Rho * sigma_A * sigma_B

portfolio_std = np.sqrt(portfolio_variance)
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(W, Rho, portfolio_std, cmap='viridis', edgecolor='none')
ax.set_xlabel('Weight of Asset A')
ax.set_ylabel('Correlation ()')
ax.set_zlabel('Portfolio Std. Deviation ()')
ax.set_title('Portfolio Standard Deviation Surface')
fig.colorbar(surf, ax=ax, shrink=0.5, aspect=10)
plt.show()
```
